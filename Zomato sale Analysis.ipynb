# DataCleaning

# Rate

```python
df['rate'].unique()
```

```python
def rate_handler(value):
    if value=='NEW' or value=="-":
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']=df['rate'].apply(rate_handler)
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

# Fill Null Values

```python
df['rate']=df['rate'].fillna(df['rate'].mean())
df['rate'].isnull().sum()
```

```python
df.info()
```

# Drop Null Values

```python
df.isnull().sum().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

# Rename Column

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)
```

```python
df.head()
```

# Drop Same Feature Attribute

```python
df.drop('city',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['cost2plates'].unique()
```

```python
def comma_handler(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
    
    
df['cost2plates']=df['cost2plates'].apply(comma_handler)
print(df['cost2plates'].unique())
print('Type:',df['cost2plates'].dtype)
```

```python
df.head()
```

# Cleaning Rest_Type

```python
df['rest_type']
```

```python
df['rest_type'].unique()
```

```python
df['rest_type'].nunique()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_less_1000=rest_type[rest_type<1000]
rest_type_less_1000
```

```python
def handle_rest_type(value):
    if value in rest_type_less_1000:
        return 'other'
    else:
        return value
    

df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].head()
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

# Location

```python
df['location']
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts()

location_less300=location[location<300]

def handle_location(value):
    if value in location_less300:
        return 'other'
    else:
        return value
    
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

# Cuisine

```python
cuisines=df['cuisines'].value_counts()
```

```python
cuisines
```

```python
cuisines_less_100=cuisines[cuisines<100]
```

```python
cuisines_less_100
```

```python
def handle_cuisines(value):
    if value in cuisines_less_100:
        return 'others'
    else:
        return value
    
    
    
    
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
